```python
from alpha_vantage.timeseries import TimeSeries
import json
import pandas as pd
class ScriptData:
    data_dict = {}
    
    def __init__(self):
        self.ts = TimeSeries(key='YXYO83L4QEYC16GP')
    
    def fetch_intraday_data(self,script):
        self.data_dict[script] = (self.ts.get_intraday(script))[0]
    
    def convert_intraday_data(self,script):
        df = pd.read_json(json.dumps(self.data_dict[script],indent = 4))
        self.data_dict[script] = df
        
    def __getitem__(self,key):
        df = self.data_dict[key].transpose()
        df.columns = ['open','high','low','close','volume']
        df.index.name = 'timestamp'
        df.reset_index(inplace = True)
        return df
        
    def __setitem__(self,key,value):
        self.data_dict[key] = value
        
    def __contains__(self,key):
        if key in self.data_dict.keys():
            return True
        return False

        
```

```python
script_data = ScriptData()
```

```python
script_data.fetch_intraday_data('GOOGL')
script_data.convert_intraday_data('GOOGL')
script_data['GOOGL']
```

```python
script_data.fetch_intraday_data('AAPL')
script_data.convert_intraday_data('AAPL')
script_data['AAPL']
```

# function - indicator1

```python
def indicator1(df, timeperiod) :
    df1 = pd.DataFrame(columns = ['timestamp','indicator'])
    
    for index,row in df.iterrows():
        sum = 0
        if (index+1) % timeperiod == 0:
            sum += row['close']
            avg = sum / timeperiod
            new_row = {'timestamp':row['timestamp'], 'indicator':avg}
            sum = 0
        else:
            sum += row['close']
            new_row = {'timestamp':row['timestamp']}
            
        df1 = df1.append(new_row,ignore_index = True)
        
    print(df1)
```

```python
indicator1(script_data['GOOGL'],timeperiod = 5)
```
